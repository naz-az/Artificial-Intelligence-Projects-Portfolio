# Dataset Description:

## Methodology:
The code follows a systematic methodology for building and training regression models to predict housing prices. The steps involved are as follows:

1. **Data Loading**: The code reads the train and test datasets from CSV files using the pandas' `read_csv` function.

2. **Data Preprocessing**:

4. **Model Building and Training**:

5. **Model Evaluation**:

6. **Prediction Visualization**:

The following code loads the train and test datasets from CSV files.



```python
dftrain=pd.read_csv('/kaggle/input/playground-series-s3e6/train.csv')
dftest =pd.read_csv('/kaggle/input/playground-series-s3e6/test.csv')
```

The following code drops the 'id' and 'cityCode' columns from the train and test datasets. This step is performed to remove unnecessary columns that may not contribute to the model's prediction.

```python
dftrain=dftrain.drop(['id','cityCode'], axis=1)
dftest=dftest.drop(['id','cityCode'], axis=1)

dftrain.shape
```

The code creates a copy of the train dataset and assigns it to a new variable 'df'. This is done to preserve the original dataset while performing further operations.

```python
df=dftrain.copy()
```

**Duplicates and Null values**

The code block following that provides some exploratory data analysis. It prints the number of data points in the dataset, checks for duplicates, and identifies missing values in the DataFrame. It also replaces empty values with NaN and rechecks the missing values after replacement.

```python
import numpy as np

# How many reviews do we have?
print('There are', df.shape[0], 'data in this dataset')

# Do we have duplicates?
print('Number of Duplicates:', len(df[df.duplicated()]))

# Do we have missing values?
missing_values = df.isnull().sum()
print('Number of Missing Values by column:\n',missing_values)

print('Number of Missing Values:', df.isnull().sum().sum())

df.replace("", np.nan, inplace=True)
missing_values = df.isnull().sum()
print('Number of Missing Values by column after replacement:\n',missing_values)
print('Number of Missing Values after replacement:', df.isnull().sum().sum())
```

The code selects the first 15 columns of features from the dftrain DataFrame and stores them in the 'features_train' variable. The 'sf' variable is used to scale down the 'label_train' values.

```python
features_train=dftrain.iloc[:,0:15]
sf=100000
label_train=dftrain.iloc[:,15]/sf
```

The next set of codes prepare the train and test features for modeling. It scales the features using standardization (standard scaling) to ensure that they are on a similar scale. This preprocessing step is important for many machine learning algorithms.

```python
scaler=stdscl()
features_train= scaler.fit_transform(features_train)
dftest= scaler.transform(dftest)
```

The code splits the train dataset into training and validation sets, as well as the corresponding labels. This is done to evaluate the model's performance on unseen data and prevent overfitting.

```python
train,val,label_train,label_val=tts(features_train,label_train,test_size=0.15,random_state=42)
```

```python
train.shape, val.shape, label_train.shape,label_val.shape
```

A callback function is defined to save the model weights with the lowest validation mean absolute error (MAE). This is useful for tracking the best model during training and ensuring that it can be used later.

```python
#define a callback to save the model weights with the lowest validation MAE
checkpoint_path = "best_weights.h5"
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    checkpoint_path,
    monitor='val_mae',
    save_best_only=True,
    mode='min',
    verbose=1)
```

```python
train.shape[1]
```

**Model**

The code builds a neural network model using the Keras Sequential API. It defines several dense (fully connected) layers with different activation functions and batch normalization layers. The model is compiled with the mean absolute error (MAE) loss function, Adam optimizer, and MAE metric. The model is trained using the training data and validated using the validation data. It runs for 250 epochs with a batch size of 250. The checkpoint callback ensures that the model weights are saved only when the validation MAE is minimized.

```python
# Build the neural network
tf.random.set_seed(42)

# Create a model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_dim=train.shape[1], activation='sigmoid'),  # Layer 1
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(128, activation='sigmoid'),  # Layer 2
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='sigmoid'),   # Layer 3
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='sigmoid'),   # Layer 4
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='sigmoid'),   # Layer 5
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='sigmoid'),   # Layer 6
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(16, activation='sigmoid'),   # Layer 7
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(1)                           # Layer 8 (output layer)
])

# Compile the model
model.compile(loss='mae',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['mae'])

# Fit the model
history = model.fit(train, label_train,
                    validation_data=(val, label_val),
                    epochs=250, batch_size=250,
                    callbacks=[checkpoint_callback],
                    verbose=1)
```

After training, the best model weights are loaded from the saved checkpoint file. The code then evaluates the best model on the validation set, calculating the test loss (MAE).

```python
#load the best model weights
model.load_weights(checkpoint_path)

#evaluate the best model on the test set
test_loss = model.evaluate(val, label_val)
```

```python
model.summary()
```

The code then plots the training and validation loss curves (history) over the epochs. This visualization helps understand the model's learning progress and identify any overfitting or underfitting patterns.

```python
#plot history (Also known as loss  curve or a training curve)
pd.DataFrame(history.history).plot()
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.ylim([0,50])
```


The following code defines a callback to save the model weights with the lowest validation mean absolute error (MAE).

```python
#define a callback to save the model weights with the lowest validation MAE
checkpoint_path2 = "best_weights2.h5"
checkpoint_callback2 = tf.keras.callbacks.ModelCheckpoint(
    checkpoint_path2,
    monitor='val_mae',
    save_best_only=True,
    mode='min',
    verbose=1)
```

**Model 2**

The next code block builds a new neural network model. The random seed is set to ensure reproducibility. The model is created using the tf.keras.Sequential() function, which defines a sequence of layers. The model consists of eight dense layers with ReLU activation and batch normalization layers after each dense layer. The last layer is the output layer with a single neuron. The model is compiled with the mean absolute error (MAE) loss function, Adam optimizer, and MAE metric. The model is then fit to the training data and validated on the validation data for 250 epochs with a batch size of 250. The ModelCheckpoint callback is used to save the weights of the best model based on the validation MAE.

```python
# Build the neural network
tf.random.set_seed(42)

# Create a model
model2=tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_dim=train.shape[1], activation='relu'),  # Layer 1
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(128, activation='relu'),  # Layer 2
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='relu'),   # Layer 3
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='relu'),   # Layer 4
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='relu'),   # Layer 5
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='relu'),   # Layer 6
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(16, activation='relu'),   # Layer 7
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(1)                           # Layer 8 (output layer)
])

# Compile the model
model2.compile(loss='mae',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['mae'])

# Fit the model
history2 = model2.fit(train, label_train,
                    validation_data=(val, label_val),
                    epochs=250, batch_size=250,
                    callbacks=[checkpoint_callback2],
                    verbose=1)
```

The best model weights are loaded from the checkpoint file. The best model is then evaluated on the validation data, and the test loss is calculated.

```python
#load the best model weights
model2.load_weights(checkpoint_path2)

#evaluate the best model on the test set
test_loss = model2.evaluate(val, label_val)
```

The training history is plotted using the loss values obtained during training, displaying the loss curve over the epochs. The y-axis represents the loss, and the x-axis represents the number of epochs. The range of the y-axis is set to [0, 50].

```python
#plot history (Also known as loss  curve or a training curve)
pd.DataFrame(history2.history).plot()
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.ylim([0,50])
```

**Visualizing Predictions**

The code defines a function called 'predict_random_sample' which takes the trained model, validation features, validation labels, and a scaler as inputs. It randomly selects an index from the validation dataset and retrieves the corresponding feature sample and actual label price. It then predicts the price using the model and calculates the difference and percentage difference between the actual and predicted prices. The original features of the sample are also displayed.

```python
def predict_random_sample(model, val_features, val_labels, scaler):
    index = np.random.randint(0, len(val_labels))
    random_sample = val_features[index]
    actual_price = val_labels.iloc[index] * sf
    prediction = model.predict(np.expand_dims(random_sample, axis=0))[0][0] * sf

    difference = abs(actual_price - prediction)
    percentage_difference = (difference / actual_price) * 100

    print("Actual Price:", actual_price)
    print("Predicted Price:", prediction)
    print("Difference:", difference)
    print("Percentage Difference:", percentage_difference, "%")

    original_features = scaler.inverse_transform(np.expand_dims(random_sample, axis=0))
    print("Features:")
    for i, feature in enumerate(dftrain.columns[:-1]):
        print(f"{feature}: {original_features[0][i]}")
```

```python
predict_random_sample(model, val, label_val, scaler)
```
