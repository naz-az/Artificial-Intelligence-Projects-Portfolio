**Evaluation Metrics**

The next code block defines two utility functions: mean_absolute_scaled_error() and evaluate_preds(). The mean_absolute_scaled_error() function calculates the Mean Absolute Scaled Error (MASE), assuming no seasonality of the data. The evaluate_preds() function calculates various metrics, including MAE, MSE, RMSE, MAPE, MASE, and R-squared score, for evaluating the predictions.

```python
# MASE implemented courtesy of sktime - https://github.com/alan-turing-institute/sktime/blob/ee7a06843a44f4aaec7582d847e36073a9ab0566/sktime/performance_metrics/forecasting/_functions.py#L16
def mean_absolute_scaled_error(y_true, y_pred):
  """
  Implement MASE (assuming no seasonality of data).
  """
  mae = tf.reduce_mean(tf.abs(y_true - y_pred))

  # Find MAE of naive forecast (no seasonality)
  mae_naive_no_season = tf.reduce_mean(tf.abs(y_true[1:] - y_true[:-1])) # our seasonality is 1 day (hence the shifting of 1 day)

  return mae / mae_naive_no_season

def evaluate_preds(y_true, y_pred):
    # Make sure float32 (for metric calculations)
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)

    # Calculate various metrics
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
    rmse = tf.sqrt(mse)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
    mase = mean_absolute_scaled_error(y_true, y_pred)

    # Calculate R-squared score
    ss_res = tf.reduce_sum((y_true - y_pred)**2)
    ss_tot = tf.reduce_sum((y_true - tf.reduce_mean(y_true))**2)
    r2 = 1 - (ss_res / ss_tot)

    # Account for different sized metrics
    if mae.ndim > 0:
        mae = tf.reduce_mean(mae)
        mse = tf.reduce_mean(mse)
        rmse = tf.reduce_mean(rmse)
        mape = tf.reduce_mean(mape)
        mase = tf.reduce_mean(mase)
        r2 = tf.reduce_mean(r2)

    return {"mae": mae.numpy(),
            "mse": mse.numpy(),
            "rmse": rmse.numpy(),
            "mape": str(mape.numpy()) + "%",
            "mase": mase.numpy(),
            "r2": r2.numpy()}
```

**Train-Test Split**

The data is split into training and testing sets using the train_test_split() function from scikit-learn.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

The next code block separates the numerical and categorical columns from the DataFrame. The numerical columns are stored in the num_cols list, while the categorical columns are stored in the cat_cols list. The lists are then printed.

```python
# Get list of numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove 'price' from num_cols
num_cols.remove(target)

# Get list of categorical columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Print the numerical columns
print("Numerical columns are:", num_cols)

# Print the categorical columns
print("Categorical columns are:", cat_cols)
```

A column transformer is created using the make_column_transformer() function from scikit-learn. This transformer applies MinMaxScaler to normalize the numerical columns and OneHotEncoder to encode the categorical columns, ignoring unknown categories.

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# creating column transformer (this will help us normalize/preprocess our data)
ct = make_column_transformer((MinMaxScaler(), num_cols),
                             (OneHotEncoder(handle_unknown = 'ignore'), cat_cols))
```

The column transformer is fitted on the training data using the fit() method, and then the training and test data are transformed using the transform() method.

```python
# fitting column transformer on training data
ct.fit(X_train)

# transforming training and test data with normalizing (MinMaxScaler) and one hot encoding (OneHotEncoder)
X_train = ct.transform(X_train)
X_test = ct.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, BatchNormalization, MaxPooling1D, Bidirectional, LSTM, Dropout, Dense, Lambda
from tensorflow.keras.optimizers import Adam
```

The transformed data is in sparse format, so it is converted to dense format using the toarray() method. Similarly, the target variables (y_train and y_test) are converted to NumPy arrays using the values attribute.

```python
X_train=X_train.toarray()
X_test=X_test.toarray()

y_train=y_train.values
y_test=y_test.values
```

**Model: Dense + Learning Rate Scheduler**

The next code block defines a neural network model using the TensorFlow Keras Sequential API. The model consists of multiple dense layers with ReLU activation and dropout regularization. The last layer has a single unit since it's a regression task.

```python
from tensorflow.keras import regularizers

# Set random seed
tf.random.set_seed(42)

# Define the model
model_1 = tf.keras.Sequential([
  tf.keras.layers.Dense(512, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])
```

The model is compiled with the mean squared error (MSE) loss function and the Adam optimizer. The learning rate is scheduled using a learning rate scheduler callback. The model is then trained on the training data, with validation performed on the test data.

```python
# Define the optimizer
optimizer = tf.keras.optimizers.Adam()

# Compile the model
model_1.compile(loss="mean_squared_error",
                optimizer=optimizer,
                metrics=["mean_absolute_error"])

# Define the learning rate schedule
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-6 * 10**(-epoch / 20))

# Fit the model
history = model_1.fit(X_train,
                      y_train,
                      epochs=100,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=[lr_schedule])
```

A plot is generated to visualize the learning rate versus the loss to aid in selecting an optimal learning rate.

```python
# Plot the learning rate versus the loss
lrs = 1e-6 * (10 ** (np.arange(100)/20))
plt.figure(figsize=(10, 7))
plt.semilogx(lrs, history.history["loss"]) # we want the x-axis (learning rate) to be log scale
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning rate vs. loss");
```

The code block resets the states generated by Keras to ensure a clean state before training a new model. The next line of code assigns a learning rate value of 1e-3 based on the observation above to the variable lrx.

```python
# Reset states generated by Keras
tf.keras.backend.clear_session()

lrx=1e-3
lrx
```

The following code block imports the necessary libraries and defines two callbacks: ModelCheckpoint and EarlyStopping.The code then defines the optimizer for the model as Adam with a learning rate specified by lrx. The model is compiled with the mean squared error (MSE) loss function and mean absolute error (MAE) as the metric. The model is trained on the training data for a specified number of epochs, with validation performed on the test data. The callbacks are passed to the callbacks parameter to save the best model and perform early stopping if needed.

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', save_best_only=True, verbose=1)
ES=EarlyStopping(patience=150, restore_best_weights=True),

# Define the optimizer
optimizer = tf.keras.optimizers.Adam(lr=lrx)

# Compile the model
model_1.compile(loss="mean_squared_error",
                optimizer=optimizer,
                metrics=["mean_absolute_error"])

# Fit the model
history = model_1.fit(X_train,
                      y_train,
                      epochs=1000,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=[ES,checkpoint])
```

**Model: Dense2**

A new neural network model called model_dense2 is defined using the TensorFlow Keras Sequential API. It consists of multiple dense layers with ReLU activation. The model is compiled with the mean absolute error (MAE) loss function and the RMSprop optimizer with a specific learning rate. MAE is chosen as the loss function because it provides a more robust measure of error compared to MSE in the presence of outliers. The model is trained on the training data for a specified number of epochs, with validation performed on the test data. The cc callback is passed to the callbacks parameter to save the best model based on the validation loss.

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

cc = ModelCheckpoint('best_model_dense2.h5',
                             monitor='val_loss',
                             mode='min',
                             save_best_only=True,
                             verbose=1)
```

```python
# Set random seed
tf.random.set_seed(42)

# Build the model
model_dense2 = tf.keras.Sequential([
  tf.keras.layers.Dense(512, activation='relu'),
  tf.keras.layers.Dense(256, activation='relu'),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_dense2.compile(loss=tf.keras.losses.mae,
                    optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
                    metrics=['mae'])

# Fit the model
model_dense2.fit(X_train, y_train, epochs=1000, validation_data=(X_test, y_test), verbose=1, callbacks=[cc])
```

The code block creates a DataFrame (X_df) from the training data (X_train) and performs label encoding for categorical columns. This is done to convert categorical variables into numeric format so that they can be used as input for certain models that require numerical data.

```python
X_df = pd.DataFrame(X_train)

# Now you can use `select_dtypes`:
for colname in X_df.select_dtypes(["category"]):
    X_df[colname] = X_df[colname].cat.codes
```

The code block imports various regressor models from scikit-learn and creates a dictionary (models) to store the models.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

models = {}

models["KNeighborsRegressor"] = {"model": KNeighborsRegressor(n_neighbors=50)}
models["LinearRegression"] = {"model": LinearRegression()}
models["XGBRegressor"] = {"model": XGBRegressor(n_jobs=5, learning_rate=0.1, max_depth=10, random_state=1)}
models["CatBoostRegressor"]={"model":CatBoostRegressor(logging_level ='Silent',iterations=500,random_state=1)}
```

The score_dataset() function is defined to evaluate a model's performance using cross-validation and return the R-squared score. The function converts categorical variables to numeric codes before scoring.

```python
from sklearn.model_selection import cross_val_score
def score_dataset(X, y, model=XGBRegressor()):
    for colname in X.select_dtypes(["category"]):
        X[colname] = X[colname].cat.codes

    score_r2 = cross_val_score(model, X, y, cv=5, scoring="r2")
    score_r2 = score_r2.mean()

    return score_r2
```

The get_scores() function is defined to iterate over the models in the models dictionary, fit each model on the training data, and evaluate its performance using the score_dataset() function.

```python
 def get_scores(models,xtrain,ytrain):
    for name,model in models.items():
        model["model"].fit(xtrain,ytrain)

        score_r2 = score_dataset(xtrain, ytrain, model=model["model"])
        score = {'model':"Linear regression", 'score_r2':score_r2}
        print("--- "+name+" ---")
        print("Score r2: {}".format(score_r2))
        print("\n")
```

The get_scores() function is called to evaluate and print the R-squared scores of the models.

```python
get_scores(models,X_df,y_train)
```
