# Methodology:

The methodology of this Eminem lyric generation project can be summarized as follows:

1. Data Acquisition: The lyrics dataset is obtained from Genius.com, a reputable source for music lyrics. It contains information such as album names, song names, lyrics, album URLs, song views, and release dates. The dataset is in CSV format.

2. Data Preprocessing: The lyrics data undergoes cleaning steps using regular expressions. Square brackets and their contents are removed, slashes are replaced with spaces, and the lyrics are split into separate lines. Empty lines are removed, resulting in cleaned lyrics ready for further processing.

3. Text Tokenization and Sequencing: The cleaned lyrics are converted to lowercase and tokenized using the Tokenizer object from Keras. Tokenization involves splitting the text into individual words or tokens. The sequences of tokens are generated by sliding a window over the text, resulting in n-gram sequences.

4. Padding: The generated n-gram sequences are padded to a consistent length using the pad_seqs function. Padding ensures that all input sequences have the same length, a requirement for model training.

5. Feature and Label Generation: The padded n-gram sequences are split into features and labels. The features are the input sequences without the last word, and the labels are the last words of the input sequences. The labels are one-hot encoded for training the model.

6. Model Creation and Training: A text generator model is created using the Keras Sequential API. The model consists of an Embedding layer, a Bidirectional LSTM layer, and a Dense layer with a softmax activation function. The model is compiled with appropriate loss and optimization functions. The model is then trained using the features and labels generated from the dataset.

7. Model Evaluation: The training progress is visualized by plotting the training accuracy and loss over the epochs. This evaluation helps in assessing the model's performance and identifying any overfitting or underfitting tendencies.

8. Text Generation: Using the trained model, text generation is performed by providing a seed text and specifying the number of words to generate. The model predicts the next word based on the seed text and continues generating subsequent words until the desired number of words is reached. The generated text reflects the style and wordplay characteristic of Eminem's lyrics.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
```

**Text Cleaning**

Now we will perform data preprocessing on the lyrics. The code initializes an empty list called cleaned_lyrics_list to store the processed lyrics. It then iterates over each song's lyrics in the lyrics_list and applies several cleaning steps using regular expressions. Square brackets and their contents are removed, slashes are replaced with spaces, and the lyrics are split into separate lines. Empty lines are also removed. The cleaned lyrics are then added to the cleaned_lyrics_list.

```python
import re

cleaned_lyrics_list = []
for song_lyrics in lyrics_list:
    # remove square brackets and their contents
    no_brackets = re.sub(r'\[.*?\]', '', song_lyrics)
    # replace slashes with spaces
    no_slashes = re.sub(r'/', ' ', no_brackets)
    # split the cleaned lyrics into separate lines
    lines = no_slashes.split('\n')
    # remove any remaining empty lines
    lines = [line for line in lines if line.strip() != '']
    cleaned_lyrics_list.extend(lines)

cleaned_lyrics_list[:3]
```

The next code reduces the size of the dataset by keeping only the first 1/4 of the cleaned lyrics list. This is done by slicing the cleaned_lyrics_list using list indexing.

```python
corp = cleaned_lyrics_list[:len(cleaned_lyrics_list)//4]
```

The code converts the cleaned lyrics into lowercase and saves them as a list called corpus. The code then prints the total number of lines in the corpus and displays the first five lines for inspection. This helps in understanding the structure and content of the text data.

```python
# Convert to lowercase and save as a list
corpus = [line.lower() for line in corp]

print(f"There are {len(corpus)} lines of text\n")
print(f"The first 5 lines look like this:\n")
print(corpus[:20])
```

Next, a tokenizer object is created using the Tokenizer() function. This tokenizer will be used to tokenize the text into individual words or tokens for further processing. The total number of unique words in the corpus is determined by adding 1 to the length of the tokenizer's word index.

**Generating n_grams**

The code defines a function called "n_gram_seqs" that generates n-gram sequences from a corpus of text. It takes the "corpus" and the tokenizer as inputs and returns a list of n-gram sequences.

```python
def n_gram_seqs(corpus, tokenizer):
    """
    Generates a list of n-gram sequences

    Args:
        corpus (list of string): lines of texts to generate n-grams for
        tokenizer (object): an instance of the Tokenizer class containing the word-index dictionary

    Returns:
        input_sequences (list of int): the n-gram sequences for each line in the corpus
    """
    input_sequences = []

    for line in corpus:
      token_list = tokenizer.texts_to_sequences([line])[0]
      for i in range(1, len(token_list)):
          n_gram_sequence = token_list[:i+1]
          input_sequences.append(n_gram_sequence)

    return input_sequences
```

The function is tested with an example by generating n-gram sequences for the first line of the "corpus" and printing the result.

```python
# Test the function with one example
first_example_sequence = n_gram_seqs([corpus[0]], tokenizer)

print("n_gram sequences for first example look like this:\n")
first_example_sequence
```

The function is further tested by generating n-gram sequences for the next three lines of the "corpus" and displaying the result.

```python
# Test the function with a bigger corpus
next_3_examples_sequence = n_gram_seqs(corpus[1:4], tokenizer)

print("n_gram sequences for next 3 examples look like this:\n")
next_3_examples_sequence
```

The code applies the "n_gram_seqs" transformation to the entire "corpus" to generate n-gram sequences for the entire corpus. It also calculates the maximum length of the sequences and prints the number of n-gram sequences and the maximum sequence length.

```python
# Apply the n_gram_seqs transformation to the whole corpus
input_sequences = n_gram_seqs(corpus, tokenizer)

# Save max length
max_sequence_len = max([len(x) for x in input_sequences])

print(f"n_grams of input_sequences have length: {len(input_sequences)}")
print(f"maximum length of sequences is: {max_sequence_len}")
```

The code defines a function called "pad_seqs" that pads tokenized sequences to the same length. It takes the input sequences and the maximum sequence length as inputs and returns the padded sequences.

```python
def pad_seqs(input_sequences, maxlen):
    """
    Pads tokenized sequences to the same length

    Args:
        input_sequences (list of int): tokenized sequences to pad
        maxlen (int): maximum length of the token sequences

    Returns:
        padded_sequences (array of int): tokenized sequences padded to the same length
    """
    padded_sequences = pad_sequences(input_sequences, maxlen=maxlen, padding='pre')

    return padded_sequences
```

The function is tested with the n-gram sequence of the first example by padding it to the maximum length and printing the result.

```python
# Test the function with the n_grams_seq of the first example
first_padded_seq = pad_seqs(first_example_sequence, max([len(x) for x in first_example_sequence]))
first_padded_seq
```

The function is further tested with the n-gram sequences of the next three examples by padding them to the maximum length and displaying the result.

```python
# Test the function with the n_grams_seq of the next 3 examples
next_3_padded_seq = pad_seqs(next_3_examples_sequence, max([len(s) for s in next_3_examples_sequence]))
next_3_padded_seq
```

The code applies the "pad_seqs" function to the entire "corpus" to pad all the n-gram sequences to the maximum length. It prints the shape of the padded corpus.

```python
# Pad the whole corpus
input_sequences = pad_seqs(input_sequences, max_sequence_len)

print(f"padded corpus has shape: {input_sequences.shape}")
```

The code defines a function called "features_and_labels" that generates features and labels from n-gram sequences. It takes the input sequences and the total number of words as inputs and returns arrays of features and one-hot encoded labels.

```python
def features_and_labels(input_sequences, total_words):
    """
    Generates features and labels from n-grams

    Args:
        input_sequences (list of int): sequences to split features and labels from
        total_words (int): vocabulary size

    Returns:
        features, one_hot_labels (array of int, array of int): arrays of features and one-hot encoded labels
    """
    input_sequences = np.array(input_sequences)
    features, labels = input_sequences[:,:-1], input_sequences[:,-1]
    one_hot_labels = to_categorical(labels, num_classes=total_words)

    return features, one_hot_labels
```

The function is tested with the padded n-gram sequence of the first example by generating features and labels and printing their shapes. It also displays the features.

```python
# Test the function with the padded n_grams_seq of the first example
first_features, first_labels = features_and_labels(first_padded_seq, total_words)

print(f"labels have shape: {first_labels.shape}")
print("\nfeatures look like this:\n")
first_features
```

The function is further tested with the entire padded corpus by generating features and labels and printing their shapes.

```python
# Split the whole corpus
features, labels = features_and_labels(input_sequences, total_words)

print(f"features have shape: {features.shape}")
print(f"labels have shape: {labels.shape}")
```

The code defines a function called "create_model" that creates a text generator model using the Keras Sequential API. It takes the total number of words and the maximum sequence length as inputs and returns the model.

```python
def create_model(total_words, max_sequence_len):
    """
    Creates a text generator model

    Args:
        total_words (int): size of the vocabulary for the Embedding layer input
        max_sequence_len (int): length of the input sequences

    Returns:
        model (tf.keras Model): the text generator model
    """
    model = Sequential()

    model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
    model.add(Bidirectional(LSTM(150, return_sequences=False)))
    model.add(Dense(total_words, activation='softmax'))


    # Compile the model
    model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])


    return model
```

The code gets an untrained model by calling the "create_model" function with the total number of words and maximum sequence length as arguments. The model is trained using the features and labels generated from the entire corpus for 100 epochs. The training progress is displayed.

```python
# Get the untrained model
model = create_model(total_words, max_sequence_len)

# Train the model
history = model.fit(features, labels, epochs=100, verbose=1)
```

The code visualizes the training curves of the model by plotting the training accuracy and training loss over the epochs.

```python
# Take a look at the training curves of the model

acc = history.history['accuracy']
loss = history.history['loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training accuracy')
plt.title('Training accuracy')

plt.figure()

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.title('Training loss')
plt.legend()

plt.show()
```

The code sets a seed text and the number of words to generate. It uses the trained model to generate the specified number of words based on the given seed text. The generated text is then printed.

```python
seed_text = "Dear Bob, i want to tell you something"
next_words = 500

for _ in range(next_words):
    # Convert the text into sequences
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    # Pad the sequences
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    # Get the probabilities of predicting a word
    predicted = model.predict(token_list, verbose=0)
    # Choose the next word based on the maximum probability
    predicted = np.argmax(predicted, axis=-1).item()
    # Get the actual word from the word index
    output_word = tokenizer.index_word[predicted]
    # Append to the current text
    seed_text += " " + output_word

print(seed_text)
```
