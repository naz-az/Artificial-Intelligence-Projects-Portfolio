# Methodology:

The DeepDrake project follows a systematic methodology to generate Drake-inspired lyrics:

1. **Data Preprocessing**: The initial step involves loading the dataset, checking for duplicates and missing values, and cleaning the lyrics by removing square brackets, slashes, and empty lines. This ensures that the dataset is ready for further analysis and model training.

2. **Text Tokenization**: The cleaned lyrics are tokenized, splitting them into individual words or phrases. This process helps in representing the lyrics as sequences of numerical tokens, making it suitable for input to the deep learning model.

3. **Sequence Generation**: The tokenized lyrics are then used to generate sequences of subphrases. By iteratively creating subphrases of increasing lengths, the model learns the contextual relationships between words and improves its ability to predict the next word in a sequence.

4. **Model Building**: A sequential deep learning model is constructed using the Keras library. The model architecture includes an embedding layer, a bidirectional LSTM layer, and a dense layer with softmax activation. This design enables the model to capture the semantic and syntactic patterns in the lyrics and generate coherent sequences.

5. **Model Training**: The model is trained using the input sequences and corresponding labels. The training process involves optimizing the model's parameters using the categorical cross-entropy loss function and the Adam optimizer. The model undergoes multiple epochs to refine its predictive capabilities.

6. **Model Evaluation**: The model's performance is evaluated by analyzing its accuracy over the training epochs. Visualizations, such as accuracy graphs, are generated using Matplotlib to assess the model's learning progress.

7. **Lyrics Generation**: Finally, the trained model is utilized to generate new lyrics. Starting from a seed text, the model predicts the next word based on the context and probability distribution. The process continues iteratively, generating a sequence of predicted words to form the generated lyrics.

The code converts the lyrics column into a list.

```python
lyrics_list = df['lyrics'].tolist()
lyrics_list[:3]
```

It then cleans the lyrics by removing square brackets and their contents, replacing slashes with spaces, splitting the cleaned lyrics into separate lines, and removing any remaining empty lines.

```python
import re

cleaned_lyrics_list = []
for song_lyrics in lyrics_list:
    # remove square brackets and their contents
    no_brackets = re.sub(r'\[.*?\]', '', song_lyrics)
    # replace slashes with spaces
    no_slashes = re.sub(r'/', ' ', no_brackets)
    # split the cleaned lyrics into separate lines
    lines = no_slashes.split('\n')
    # remove any remaining empty lines
    lines = [line for line in lines if line.strip() != '']
    cleaned_lyrics_list.extend(lines)

cleaned_lyrics_list[:3]
```

The next code reduces the size of the dataset by keeping only the first 1/4 of the cleaned lyrics list. This is done by slicing the cleaned_lyrics_list using list indexing.

```python
cleaned_lyrics_list = cleaned_lyrics_list[:len(cleaned_lyrics_list)//4]
```

The following code shows how to write a list of cleaned lyrics to a file named 'lyrics.txt' using a loop and file writing operations.

```python
# Now you can write it to a file
with open('lyrics.txt', 'w') as file:
    for line in cleaned_lyrics_list:
        file.write(str(line) + '\n')
```

The next code segment loads the contents of the 'lyrics.txt' file, converts the text to lowercase, and splits it into a list of sentences. This preprocessing step prepares the text data for further analysis.

```python
# Load the dataset
data = open('/kaggle/working/lyrics.txt').read()

# Lowercase and split the text
corpus = data.lower().split("\n")

# Preview the result
print(corpus[:3])
```

Now we will perform tokenization and generate a word index dictionary using the Tokenizer class. The word index dictionary maps each unique word in the corpus to a numerical index. Additionally, we calculate the total number of unique words in the corpus, which will be useful for creating one-hot encoded labels.

```python
# Initialize the Tokenizer class
tokenizer = Tokenizer()

# Generate the word index dictionary
tokenizer.fit_on_texts(corpus)

# Define the total words. You add 1 for the index `0` which is just the padding token.
total_words = len(tokenizer.word_index) + 1

print(f'word index dictionary: {tokenizer.word_index}')
print(f'total words: {total_words}')
```

The following code initializes an empty list called "input_sequences" to store sequences of tokens. It loops over each line in the corpus and tokenizes the line using the word index dictionary. Then, it generates subphrases by iterating over the tokens and appends them to the input_sequences list. The maximum length of the subphrases is determined and used to pad all sequences to a uniform length.

```python
# Initialize the sequences list
input_sequences = []

# Loop over every line
for line in corpus:

	# Tokenize the current line
	token_list = tokenizer.texts_to_sequences([line])[0]

	# Loop over the line several times to generate the subphrases
	for i in range(1, len(token_list)):

		# Generate the subphrase
		n_gram_sequence = token_list[:i+1]

		# Append the subphrase to the sequences list
		input_sequences.append(n_gram_sequence)

# Get the length of the longest line
max_sequence_len = max([len(x) for x in input_sequences])

# Pad all sequences
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')

# Create inputs and label by splitting the last token in the subphrases
xs, labels = input_sequences[:,:-1],input_sequences[:,-1]

# Convert the label into one-hot arrays
ys = tf.keras.utils.to_categorical(labels, num_classes=total_words)
```

This code takes the first sentence from a corpus, splits it into words, and prints the sentence and its tokenized indices.

**Model**

The next code segment defines hyperparameters such as embedding dimension, LSTM units, and learning rate. It then builds a sequential model using these hyperparameters and prints a summary of the model architecture. The model consists of an embedding layer, a bidirectional LSTM layer, and a dense layer with softmax activation.

```python
# Hyperparameters
embedding_dim = 100
lstm_units = 150
learning_rate = 0.01

# Build the model
model = Sequential([
          Embedding(total_words, embedding_dim, input_length=max_sequence_len-1),
          Bidirectional(LSTM(lstm_units)),
          Dense(total_words, activation='softmax')
])

# Use categorical crossentropy because this is a multi-class problem
model.compile(
    loss='categorical_crossentropy',
    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
    metrics=['accuracy']
    )

# Print the model summary
model.summary()
```

The code sets the number of epochs for training the model and proceeds to train the model using the input sequences and one-hot encoded labels. The training history is stored in the "history" variable.

```python
epochs = 50

# Train the model
history = model.fit(xs, ys, epochs=epochs)
```

To visualize the accuracy of the model during training, the code includes a utility function called "plot_graphs". It plots the accuracy values from the training history using Matplotlib.

```python
import matplotlib.pyplot as plt

# Plot utility
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.show()

# Visualize the accuracy
plot_graphs(history, 'accuracy')
```

Now we define a seed text to generate new text. We set the number of words to predict and iterate until the desired length is reached. In each iteration, the seed text is converted to a token sequence, padded, and fed into the model to get the probabilities for each index. The index with the highest probability is selected as the predicted word, which is then appended to the seed text. This process continues to generate a sequence of predicted words.

```python
# Define seed text
seed_text = "sunshine this morning"

# Define total words to predict
next_words = 100

# Loop until desired length is reached
for _ in range(next_words):

	# Convert the seed text to a token sequence
	token_list = tokenizer.texts_to_sequences([seed_text])[0]

	# Pad the sequence
	token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')

	# Feed to the model and get the probabilities for each index
	probabilities = model.predict(token_list, verbose=0)

	# Get the index with the highest probability
	predicted = np.argmax(probabilities, axis=-1)[0]

	# Ignore if index is 0 because that is just the padding.
	if predicted != 0:

		# Look up the word associated with the index.
		output_word = tokenizer.index_word[predicted]

		# Combine with the seed text
		seed_text += " " + output_word

# Print the result
print(seed_text)
```

Similarly, we define another seed text and set the number of words to predict. This time, in each iteration, a random choice is made among the three words with the highest probabilities. The probabilities are sorted in ascending order, and the random choice is made from the end of the array. The selected word is appended to the seed text, and the process continues to generate a sequence of predicted words. Finally, the resulting seed text is printed, which contains the generated lyrics based on the trained model. The same seed text and generated lyrics are printed again.

```python
# Define seed text
seed_text = "sunshine this morning"

# Define total words to predict
next_words = 500

# Loop until desired length is reached
for _ in range(next_words):

	# Convert the seed text to a token sequence
  token_list = tokenizer.texts_to_sequences([seed_text])[0]

	# Pad the sequence
  token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')

	# Feed to the model and get the probabilities for each index
  probabilities = model.predict(token_list, verbose=0)

  # Pick a random number from [1,2,3]
  choice = np.random.choice([1,2,3])

  # Sort the probabilities in ascending order
  # and get the random choice from the end of the array
  predicted = np.argsort(probabilities)[0][-choice]

	# Ignore if index is 0 because that is just the padding.
  if predicted != 0:

		# Look up the word associated with the index.
	  output_word = tokenizer.index_word[predicted]

		# Combine with the seed text
	  seed_text += " " + output_word

# Print the result
print(seed_text)
```
