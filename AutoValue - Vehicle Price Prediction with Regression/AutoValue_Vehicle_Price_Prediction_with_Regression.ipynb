**Evaluation Metrics**

The next code block defines two utility functions: mean_absolute_scaled_error() and evaluate_preds(). The mean_absolute_scaled_error() function calculates the Mean Absolute Scaled Error (MASE), assuming no seasonality of the data. The evaluate_preds() function calculates various metrics, including MAE, MSE, RMSE, MAPE, MASE, and R-squared score, for evaluating the predictions.

```python
# MASE implemented courtesy of sktime - https://github.com/alan-turing-institute/sktime/blob/ee7a06843a44f4aaec7582d847e36073a9ab0566/sktime/performance_metrics/forecasting/_functions.py#L16
def mean_absolute_scaled_error(y_true, y_pred):
  """
  Implement MASE (assuming no seasonality of data).
  """
  mae = tf.reduce_mean(tf.abs(y_true - y_pred))

  # Find MAE of naive forecast (no seasonality)
  mae_naive_no_season = tf.reduce_mean(tf.abs(y_true[1:] - y_true[:-1])) # our seasonality is 1 day (hence the shifting of 1 day)

  return mae / mae_naive_no_season

def evaluate_preds(y_true, y_pred):
    # Make sure float32 (for metric calculations)
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)

    # Calculate various metrics
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
    rmse = tf.sqrt(mse)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
    mase = mean_absolute_scaled_error(y_true, y_pred)

    # Calculate R-squared score
    ss_res = tf.reduce_sum((y_true - y_pred)**2)
    ss_tot = tf.reduce_sum((y_true - tf.reduce_mean(y_true))**2)
    r2 = 1 - (ss_res / ss_tot)

    # Account for different sized metrics
    if mae.ndim > 0:
        mae = tf.reduce_mean(mae)
        mse = tf.reduce_mean(mse)
        rmse = tf.reduce_mean(rmse)
        mape = tf.reduce_mean(mape)
        mase = tf.reduce_mean(mase)
        r2 = tf.reduce_mean(r2)

    return {"mae": mae.numpy(),
            "mse": mse.numpy(),
            "rmse": rmse.numpy(),
            "mape": str(mape.numpy()) + "%",
            "mase": mase.numpy(),
            "r2": r2.numpy()}
```

**Model: Dense**

The subsequent code block defines a neural network model using TensorFlow's Keras API. The model consists of several dense layers with activation functions and dropout regularization. This model is crucial for performing regression tasks on the dataset.

```python
from tensorflow.keras import regularizers

# Set random seed
tf.random.set_seed(42)

# Define the model
model_1 = tf.keras.Sequential([
  tf.keras.layers.Dense(512, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])
```

The code block following the model definition sets up the optimizer and compiles the previously defined neural network model. This step is necessary to configure the optimization algorithm and loss function for the model.

```python
# Define the optimizer
optimizer = tf.keras.optimizers.Adam()

# Compile the model
model_1.compile(loss="mean_squared_error",
                optimizer=optimizer,
                metrics=["mean_absolute_error"])
```

Next, a learning rate schedule is defined using a lambda function, which adjusts the learning rate based on the epoch number. This schedule helps optimize the training process of the neural network.

The subsequent code block fits the defined model to the training data. It trains the model for a specified number of epochs, using a batch size and validation data.

```python
# Define the learning rate schedule
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-6 * 10**(-epoch / 20))

# Fit the model
history = model_1.fit(X_train,
                      y_train,
                      epochs=100,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=[lr_schedule])
```

Following the model training, a plot is created to visualize the learning rate versus the loss. This plot helps analyze the relationship between the learning rate and the corresponding loss during the training process.

```python
# Plot the learning rate versus the loss
lrs = 1e-6 * (10 ** (np.arange(100)/20))
plt.figure(figsize=(10, 7))
plt.semilogx(lrs, history.history["loss"]) # we want the x-axis (learning rate) to be log scale
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning rate vs. loss");
```

Following the model imports, a function is defined to score a dataset using cross-validation with a specified model. The function converts categorical variables to numeric codes and calculates the average R-squared score across multiple cross-validation folds.

```python
from sklearn.model_selection import cross_val_score
def score_dataset(X, y, model=XGBRegressor()):
    for colname in X.select_dtypes(["category"]):
        X[colname] = X[colname].cat.codes

    score_r2 = cross_val_score(model, X, y, cv=5, scoring="r2")
    score_r2 = score_r2.mean()

    return score_r2
```

Finally, a function called "get_scores" is defined to evaluate and compare the performance of different models on the training data. The function fits each model to the training data, scores the dataset using the previously defined scoring function, and prints the R-squared scores for each mode

```python
def get_scores(models,xtrain,ytrain):
    for name,model in models.items():
        model["model"].fit(xtrain,ytrain)

        score_r2 = score_dataset(xtrain, ytrain, model=model["model"])
        score = {'model':"Linear regression", 'score_r2':score_r2}
        print("--- "+name+" ---")
        print("Score r2: {}".format(score_r2))
        print("\n")
```
