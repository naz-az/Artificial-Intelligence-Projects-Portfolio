The subsequent code block imports the ImageDataGenerator class from TensorFlow's preprocessing module. It sets the directories for the training and validation sets and creates data generators with specified augmentation parameters. The generators rescale the image pixel values and perform various transformations like rotation, shifting, shearing, zooming, and flipping. The images are resized to a target size of 150x150 pixels and the labels are one-hot encoded. The batch size is set to 126.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

TRAINING_DIR = "tmp/rps-train/rps"
training_datagen = ImageDataGenerator(
      rescale = 1./255,
	    rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

VALIDATION_DIR = "tmp/rps-test/rps-test-set"
validation_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = training_datagen.flow_from_directory(
	TRAINING_DIR,
	target_size=(150,150),
	class_mode='categorical',
  batch_size=126
)

validation_generator = validation_datagen.flow_from_directory(
	VALIDATION_DIR,
	target_size=(150,150),
	class_mode='categorical',
  batch_size=126
)
```

In the next code block, a function called plot_images is defined to plot random images from a given array of images along with their corresponding labels. The function randomly selects 10 images and displays them in a grid with their predicted class labels and the true class labels.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_images(images, labels, classes):
    # Choose random images to display
    indices = np.random.choice(np.arange(images.shape[0]), size=10, replace=False)
    images, labels = images[indices], labels[indices]

    fig, axes = plt.subplots(2, 5, figsize=(15,6))

    for i, ax in enumerate(axes.flatten()):
        ax.imshow(images[i])
        ax.set_title(classes[np.argmax(labels[i])])
        ax.axis('off')

    plt.tight_layout()
    plt.show()
# Get a batch of images and their corresponding labels
images, labels = next(train_generator)

# Get class names
class_names = list(train_generator.class_indices.keys())

plot_images(images, labels, class_names)
```

The following code block defines a function called is_one_hot to check if the labels are one-hot encoded. It checks the shape and values of the labels array to determine if it satisfies the conditions for one-hot encoding.

```python
import numpy as np

def is_one_hot(labels):
    """Checks if labels are one-hot encoded"""
    return ((labels.ndim == 2) and
            (labels.shape[1] >= 2) and
            (np.all((labels == 0) | (labels == 1))) and
            (np.all(labels.sum(axis=1) == 1)))
```

The following code block defines the model architecture using the Sequential API from TensorFlow's keras.models module. It consists of several convolutional layers with pooling, followed by a flatten layer, dropout layer, and dense layers. The model summary is then printed.

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 150x150 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # The second convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The third convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fourth convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Print the model summary
model.summary()
```

In the next code block, the model is compiled with the specified loss function, optimizer, and metrics.

```python
# Set the training parameters
model.compile(loss = 'categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

The subsequent code block trains the model using the fit function. It specifies the number of epochs, steps per epoch, validation data, verbosity, and validation steps.

```python
# Train the model
history = model.fit(train_generator, epochs=25, steps_per_epoch=20, validation_data = validation_generator, verbose = 1, validation_steps=3)
```

The following code block imports matplotlib.pyplot for plotting the training and validation accuracy and loss curves. It plots the curves based on the training history obtained from the fit function.

```python
import matplotlib.pyplot as plt

# Plot the results
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()

plt.show()
```

In the next code block, the number of steps is determined based on the length of the validation generator. Then, the true labels, predicted labels, and prediction probabilities are initialized. For each batch of images and labels from the validation generator, the true labels and prediction probabilities are stored. The predicted labels are obtained by applying argmax to the prediction probabilities.

```python
# Determine the number of steps
steps = len(validation_generator)

y_true = []
y_preds = []
y_probs = []  # Store the prediction probabilities

for i, (images, labels) in enumerate(validation_generator):
    if i < steps:  # don't iterate more than the number of batches in your dataset
        # get the class labels
        for label in labels:
            y_true.append(np.argmax(label))  # we use np.argmax to get the class from one-hot encoded label

        preds = model.predict(images, verbose=1)

        # Store the prediction probabilities
        for pred in preds:
            y_probs.append(pred)

        # get the class predictions
        for pred in preds:
            y_preds.append(np.argmax(pred))  # we use np.argmax to get the class from prediction probabilities
    else:
        break
```

The subsequent code block imports the confusion_matrix function from scikit-learn and calculates the confusion matrix based on the true and predicted labels.

```python
# Check out the non-prettified confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true=y_true,
                 y_pred=y_preds)
```

The next code block calls a function called make_confusion_matrix to plot a prettier version of the confusion matrix.

```python
# Note: The following confusion matrix code is a remix of Scikit-Learn's
# plot_confusion_matrix function - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_confusion_matrix.html
# and Made with ML's introductory notebook - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
import itertools
from sklearn.metrics import confusion_matrix

# Our function needs a different name to sklearn's plot_confusion_matrix
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15):
  """Makes a labelled confusion matrix comparing predictions and ground truth labels.

  If classes is passed, confusion matrix will be labelled, if not, integer class values
  will be used.

  Args:
    y_true: Array of truth labels (must be same shape as y_pred).
    y_pred: Array of predicted labels (must be same shape as y_true).
    classes: Array of class labels (e.g. string form). If `None`, integer labels are used.
    figsize: Size of output figure (default=(10, 10)).
    text_size: Size of output figure text (default=15).

  Returns:
    A labelled confusion matrix plot comparing y_true and y_pred.

  Example usage:
    make_confusion_matrix(y_true=test_labels, # ground truth test labels
                          y_pred=y_preds, # predicted labels
                          classes=class_names, # array of class label names
                          figsize=(15, 15),
                          text_size=10)
  """
  # Create the confustion matrix
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] # normalize it
  n_classes = cm.shape[0] # find the number of classes we're dealing with

  # Plot the figure and make it pretty
  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap=plt.cm.Blues) # colors will represent how 'correct' a class is, darker == better
  fig.colorbar(cax)

  # Are there a list of classes?
  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  # Label the axes
  ax.set(title="Confusion Matrix",
         xlabel="Predicted label",
         ylabel="True label",
         xticks=np.arange(n_classes), # create enough axis slots for each class
         yticks=np.arange(n_classes),
         xticklabels=labels, # axes will labeled with class names (if they exist) or ints
         yticklabels=labels)

  # Make x-axis labels appear on bottom
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  # Set the threshold for different colors
  threshold = (cm.max() + cm.min()) / 2.

  # Plot the text on each cell
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
             horizontalalignment="center",
             color="white" if cm[i, j] > threshold else "black",
             size=text_size)
```

```python
# Make a prettier confusion matrix
make_confusion_matrix(y_true=y_true,
                      y_pred=y_preds,
                      classes=class_names,
                      figsize=(15, 15),
                      text_size=10)
```

Finally, the code block defines a function called plot_random_images that takes a trained model, a dataset (ImageDataGenerator object), and an array of class names as input. It selects random images from the dataset, predicts their labels using the model, and plots them along with their true labels. The function returns a grid plot of the random images.

```python
def plot_random_images(model, dataset, classes):
  """Picks random images, plots them and labels them with predicted and truth labels.

  Args:
    model: a trained model (trained on data similar to what's in dataset).
    dataset: an ImageDataGenerator object of (image, label) pairs.
    classes: array of class names for images.

  Returns:
    A grid plot of random images from `dataset` with predicted class labels from `model`
    as well as the truth class labels.
  """
  # Pick a batch
  images, true_labels = next(dataset)  # Use next(dataset) instead of next(iter(dataset))

  # Setup 20 random integers
  indices = np.random.randint(0, images.shape[0], size=20)

  fig, axes = plt.subplots(5, 4, figsize=(10, 10))  # We'll have 5 rows of 4 images each

  for i, ax in enumerate(axes.flat):
    # Create predictions and targets
    target_image = images[indices[i]]
    pred_probs = model.predict(target_image[np.newaxis, ...], verbose=0)

    # Check the number of unique classes
    if len(classes) > 2:  # multiclass classification
      pred_label = classes[np.argmax(pred_probs[0])]
      true_label = classes[np.argmax(true_labels[indices[i]])]
    else:  # binary classification
      pred_label = classes[int(np.round(np.max(pred_probs[0])))]
      true_label = classes[int(np.squeeze(true_labels[indices[i]]))]

    # Plot the target image
    ax.imshow((target_image * 255).astype("uint8"))  # rescale the pixel values before displaying

    # Change the color of the titles depending on if the prediction is right or wrong
    color = "green" if pred_label == true_label else "red"

    # Add xlabel information (prediction/true label)
    ax.set_xlabel("Pred: {} {:2.0f}% (True: {})".format(pred_label,
                                                       100*np.max(pred_probs),
                                                       true_label),
                 color=color)

  plt.tight_layout()
  plt.show()
```

```python
plot_random_images(model=model,
                   dataset=validation_generator,
                   classes=class_names)
```
