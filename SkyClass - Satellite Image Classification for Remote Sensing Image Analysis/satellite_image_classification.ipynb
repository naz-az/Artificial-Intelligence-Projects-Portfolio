**Generate train and test folders**

The next code block performs data preprocessing and prepares the dataset for training. The code block starts by defining the working directory and the train-test split ratio. Then, for each class in the dataset, the code retrieves the list of all images, shuffles the file list randomly, and splits it into training and testing sets based on the defined split ratio. The code creates train and test directories in the working directory and copies the images into their respective directories.

```python
import os
import shutil
import numpy as np

working_dir = '/kaggle/working/'

# Define train and test split ratio
train_ratio = 0.8

for cls in class_names:
    # Get list of all images
    src_dir = os.path.join(base_dir, cls)
    all_files = os.listdir(src_dir)

    # Randomly shuffle the file list
    np.random.shuffle(all_files)

    # Split into training and testing sets
    train_files = all_files[:int(len(all_files)*train_ratio)]
    test_files = all_files[int(len(all_files)*train_ratio):]

    # Create train and test directories
    train_dir = os.path.join(working_dir, 'Train', cls)
    test_dir = os.path.join(working_dir, 'Test', cls)
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Copy images into respective directories
    for file in train_files:
        shutil.copy(os.path.join(src_dir, file), train_dir)
    for file in test_files:
        shutil.copy(os.path.join(src_dir, file), test_dir)
```

The next code block defines a function called is_one_hot, which checks if labels are one-hot encoded. It checks the shape and values of the labels array and returns a boolean value indicating whether the labels are one-hot encoded or not.

```python
import numpy as np

def is_one_hot(labels):
    """Checks if labels are one-hot encoded"""
    return ((labels.ndim == 2) and
            (labels.shape[1] >= 2) and
            (np.all((labels == 0) | (labels == 1))) and
            (np.all(labels.sum(axis=1) == 1)))
```

**Model: Convolutional**

The following code shows the creation of a convolutional neural network (CNN) model using the TensorFlow framework. It uses the Sequential model from Keras and includes several convolutional and pooling layers. The next code compiles the previously created model. It specifies the loss function, optimizer, and evaluation metrics for the model.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense

# Create our model
model_conv = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(filters=10,
                         kernel_size=3, # can also be (3, 3)
                         activation="relu",
                         input_shape=(224, 224, 3)), # first layer specifies input shape (height, width, colour channels)
  tf.keras.layers.Conv2D(10, 3, activation="relu"),
  tf.keras.layers.MaxPool2D(pool_size=2, # pool_size can also be (2, 2)
                            padding="valid"), # padding can also be 'same'
  tf.keras.layers.Conv2D(10, 3, activation="relu"),
  tf.keras.layers.Conv2D(10, 3, activation="relu"), # activation='relu' == tf.keras.layers.Activations(tf.nn.relu)
  tf.keras.layers.MaxPool2D(2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(n_classes, activation="softmax")
])

# Compile the model
model_conv.compile(loss="categorical_crossentropy",
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])
```

**Transfer Learning**

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers
```

The code defines URLs for pre-trained models such as ResNet, InceptionV3, and MobileNetV2 using TensorFlow Hub. These pre-trained models can be used as feature extraction layers in a custom model.

```python
resnet_url = "https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4"

inception_v3_url = "https://tfhub.dev/google/tf2-preview/inception_v3/feature_vector/4"

mobilenet_v2_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

A function called create_model is defined, which takes a TensorFlow Hub URL and creates a Keras Sequential model. It downloads the pre-trained model and uses it as the feature extraction layer. It then adds a dense output layer with the specified number of classes.

```python
def create_model(model_url, num_classes=10):
  """Takes a TensorFlow Hub URL and creates a Keras Sequential model with it.

  Args:
    model_url (str): A TensorFlow Hub feature extraction URL.
    num_classes (int): Number of output neurons in output layer,
      should be equal to number of target classes, default 10.

  Returns:
    An uncompiled Keras Sequential model with model_url as feature
    extractor layer and Dense output layer with num_classes outputs.
  """
  # Download the pretrained model and save it as a Keras layer
  feature_extractor_layer = hub.KerasLayer(model_url,
                                           trainable=False, # freeze the underlying patterns
                                           name='feature_extraction_layer',
                                           input_shape=(224, 224, 3)) # define the input image shape

  # Create our own model
  model = tf.keras.Sequential([
    feature_extractor_layer, # use the feature extraction layer as the base
    layers.Dense(num_classes, activation='softmax', name='output_layer') # create our own output layer
  ])

  return model
```

The code includes a function called plot_loss_curves, which plots the training and validation loss curves from the training history. This function provides a visual representation of the model's learning progress.

```python
# Plot the validation and training data separately
def plot_loss_curves(history):
  """
  Returns separate loss curves for training and validation metrics.
  """
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(history.history['loss']))

  # Plot loss
  plt.plot(epochs, loss, label='training_loss')
  plt.plot(epochs, val_loss, label='val_loss')
  plt.title('Loss')
  plt.xlabel('Epochs')
  plt.legend()

  # Plot accuracy
  plt.figure()
  plt.plot(epochs, accuracy, label='training_accuracy')
  plt.plot(epochs, val_accuracy, label='val_accuracy')
  plt.title('Accuracy')
  plt.xlabel('Epochs')
  plt.legend();
```

A function called make_confusion_matrix is used to create a prettier visualization of the confusion matrix. It plots the confusion matrix using a heatmap and includes class labels and other visual enhancements.

```python
# Note: The following confusion matrix code is a remix of Scikit-Learn's
# plot_confusion_matrix function - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_confusion_matrix.html
# and Made with ML's introductory notebook - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
import itertools
from sklearn.metrics import confusion_matrix

# Our function needs a different name to sklearn's plot_confusion_matrix
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15):
  """Makes a labelled confusion matrix comparing predictions and ground truth labels.

  If classes is passed, confusion matrix will be labelled, if not, integer class values
  will be used.

  Args:
    y_true: Array of truth labels (must be same shape as y_pred).
    y_pred: Array of predicted labels (must be same shape as y_true).
    classes: Array of class labels (e.g. string form). If `None`, integer labels are used.
    figsize: Size of output figure (default=(10, 10)).
    text_size: Size of output figure text (default=15).

  Returns:
    A labelled confusion matrix plot comparing y_true and y_pred.

  Example usage:
    make_confusion_matrix(y_true=test_labels, # ground truth test labels
                          y_pred=y_preds, # predicted labels
                          classes=class_names, # array of class label names
                          figsize=(15, 15),
                          text_size=10)
  """
  # Create the confustion matrix
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] # normalize it
  n_classes = cm.shape[0] # find the number of classes we're dealing with

  # Plot the figure and make it pretty
  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap=plt.cm.Blues) # colors will represent how 'correct' a class is, darker == better
  fig.colorbar(cax)

  # Are there a list of classes?
  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  # Label the axes
  ax.set(title="Confusion Matrix",
         xlabel="Predicted label",
         ylabel="True label",
         xticks=np.arange(n_classes), # create enough axis slots for each class
         yticks=np.arange(n_classes),
         xticklabels=labels, # axes will labeled with class names (if they exist) or ints
         yticklabels=labels)

  # Make x-axis labels appear on bottom
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  # Set the threshold for different colors
  threshold = (cm.max() + cm.min()) / 2.

  # Plot the text on each cell
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
             horizontalalignment="center",
             color="white" if cm[i, j] > threshold else "black",
             size=text_size)
```

```python
# Make a prettier confusion matrix
make_confusion_matrix(y_true=y_true,
                      y_pred=y_preds,
                      classes=class_names,
                      figsize=(15, 15),
                      text_size=10)
```

The plot_random_images function is defined, which selects random images from a dataset, makes predictions using the provided model, and plots the images along with their predicted and true labels. This function helps visually inspect the model's predictions.

```python
def plot_random_images(model, dataset, classes):
  """Picks random images, plots them and labels them with predicted and truth labels.

  Args:
    model: a trained model (trained on data similar to what's in dataset).
    dataset: an ImageDataGenerator object of (image, label) pairs.
    classes: array of class names for images.

  Returns:
    A grid plot of random images from `dataset` with predicted class labels from `model`
    as well as the truth class labels.
  """
  # Pick a batch
  images, true_labels = next(dataset)  # Use next(dataset) instead of next(iter(dataset))

  # Setup 20 random integers
  indices = np.random.randint(0, images.shape[0], size=20)

  fig, axes = plt.subplots(5, 4, figsize=(10, 10))  # We'll have 5 rows of 4 images each

  for i, ax in enumerate(axes.flat):
    # Create predictions and targets
    target_image = images[indices[i]]
    pred_probs = model.predict(target_image[np.newaxis, ...], verbose=0)

    # Check the number of unique classes
    if len(classes) > 2:  # multiclass classification
      pred_label = classes[np.argmax(pred_probs[0])]
      true_label = classes[np.argmax(true_labels[indices[i]])]
    else:  # binary classification
      pred_label = classes[int(np.round(np.max(pred_probs[0])))]
      true_label = classes[int(np.squeeze(true_labels[indices[i]]))]

    # Plot the target image
    ax.imshow((target_image * 255).astype("uint8"))  # rescale the pixel values before displaying

    # Change the color of the titles depending on if the prediction is right or wrong
    color = "green" if pred_label == true_label else "red"

    # Add xlabel information (prediction/true label)
    ax.set_xlabel("Pred: {} {:2.0f}% (True: {})".format(pred_label,
                                                       100*np.max(pred_probs),
                                                       true_label),
                 color=color)

  plt.tight_layout()
  plt.show()
```
