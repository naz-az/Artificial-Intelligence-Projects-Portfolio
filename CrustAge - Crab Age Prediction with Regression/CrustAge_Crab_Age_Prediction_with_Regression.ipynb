**Evaluation Metrics**

The code defines two functions, mean_absolute_scaled_error and evaluate_preds, for evaluating the performance of a regression model. These functions calculate various metrics such as mean absolute error (MAE), mean squared error (MSE), root mean squared error (RMSE), mean absolute percentage error (MAPE), mean absolute scaled error (MASE), and R-squared score.

```python
# MASE implemented courtesy of sktime - https://github.com/alan-turing-institute/sktime/blob/ee7a06843a44f4aaec7582d847e36073a9ab0566/sktime/performance_metrics/forecasting/_functions.py#L16
def mean_absolute_scaled_error(y_true, y_pred):
  """
  Implement MASE (assuming no seasonality of data).
  """
  mae = tf.reduce_mean(tf.abs(y_true - y_pred))

  # Find MAE of naive forecast (no seasonality)
  mae_naive_no_season = tf.reduce_mean(tf.abs(y_true[1:] - y_true[:-1])) # our seasonality is 1 day (hence the shifting of 1 day)

  return mae / mae_naive_no_season

def evaluate_preds(y_true, y_pred):
    # Make sure float32 (for metric calculations)
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)

    # Calculate various metrics
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
    rmse = tf.sqrt(mse)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
    mase = mean_absolute_scaled_error(y_true, y_pred)

    # Calculate R-squared score
    ss_res = tf.reduce_sum((y_true - y_pred)**2)
    ss_tot = tf.reduce_sum((y_true - tf.reduce_mean(y_true))**2)
    r2 = 1 - (ss_res / ss_tot)

    # Account for different sized metrics
    if mae.ndim > 0:
        mae = tf.reduce_mean(mae)
        mse = tf.reduce_mean(mse)
        rmse = tf.reduce_mean(rmse)
        mape = tf.reduce_mean(mape)
        mase = tf.reduce_mean(mase)
        r2 = tf.reduce_mean(r2)

    return {"mae": mae.numpy(),
            "mse": mse.numpy(),
            "rmse": rmse.numpy(),
            "mape": str(mape.numpy()) + "%",
            "mase": mase.numpy(),
            "r2": r2.numpy()}
```

**Model: Dense + LearningRateScheduler**

A sequential model is defined using the Keras API with three dense layers. The model architecture includes ReLU activation, dropout regularization, and a single output node.

```python
# Set random seed
tf.random.set_seed(42)

# Define the model
model_1 = tf.keras.Sequential([
  tf.keras.layers.Dense(512, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])
```

The optimizer is defined as Adam, and the model is compiled with the mean squared error loss function and mean absolute error as the evaluation metric.

A learning rate schedule is defined using the LearningRateScheduler callback, which adjusts the learning rate during training based on the epoch number.

The model is trained on the training data for 100 epochs with a batch size of 32. The validation data is used to monitor the model's performance during training.

```python
# Define the optimizer
optimizer = tf.keras.optimizers.Adam()

# Compile the model
model_1.compile(loss="mean_squared_error",
                optimizer=optimizer,
                metrics=["mean_absolute_error"])

# Define the learning rate schedule
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-6 * 10**(-epoch / 20))

# Fit the model
history = model_1.fit(X_train,
                      y_train,
                      epochs=100,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=[lr_schedule])
```

A plot is created to visualize the relationship between the learning rate and the loss value.

```python
# Plot the learning rate versus the loss
lrs = 1e-6 * (10 ** (np.arange(100)/20))
plt.figure(figsize=(10, 7))
plt.semilogx(lrs, history.history["loss"]) # we want the x-axis (learning rate) to be log scale
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning rate vs. loss");
```

The model's states are reset, clearing any previously learned parameters and resetting the random seed. And the learning rate is set to 0.001.

```python
# Reset states generated by Keras
tf.keras.backend.clear_session()

lrx=1e-3
lrx
```

Early stopping and model checkpoint callbacks are defined to monitor the validation loss and save the best model during training.

The optimizer is updated with the specified learning rate. The model is compiled and trained for 300 epochs with the defined callbacks.

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', save_best_only=True, verbose=1)
ES=EarlyStopping(patience=150, restore_best_weights=True),

# Define the optimizer
optimizer = tf.keras.optimizers.Adam(lr=lrx)

# Compile the model
model_1.compile(loss="mean_squared_error",
                optimizer=optimizer,
                metrics=["mean_absolute_error"])

# Fit the model
history = model_1.fit(X_train,
                      y_train,
                      epochs=300,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=[ES,checkpoint])
```

**Model: Dense2**

The next code block sets up a model checkpoint callback using the ModelCheckpoint class from TensorFlow.keras. This callback saves the best model during training based on the validation loss.

```python
from tensorflow.keras.callbacks import ModelCheckpoint

cc = ModelCheckpoint('best_model_dense2.h5',
                             monitor='val_loss',
                             mode='min',
                             save_best_only=True,
                             verbose=1)
```

A sequential model called 'model_dense2' is built using the Keras API. It consists of five dense layers with increasing sizes and ReLU activation. The model is designed for regression, with a single output node. The model is compiled with the mean absolute error (MAE) loss function, RMSprop optimizer with a learning rate of 0.001, and MAE as the evaluation metric. The model is trained on the training data for 1000 epochs, with validation data used for monitoring the model's performance. The ModelCheckpoint callback is added to save the best model based on the validation loss.

```python
# Set random seed
tf.random.set_seed(42)

# Build the model
model_dense2 = tf.keras.Sequential([
  tf.keras.layers.Dense(512, activation='relu'),
  tf.keras.layers.Dense(256, activation='relu'),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_dense2.compile(loss=tf.keras.losses.mae,
                    optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
                    metrics=['mae'])

# Fit the model
model_dense2.fit(X_train, y_train, epochs=1000, validation_data=(X_test, y_test), verbose=1, callbacks=[cc])
```

The score_dataset() function is defined to evaluate a model's performance using cross-validation and return the R-squared score. The function converts categorical variables to numeric codes before scoring.

```python
from sklearn.model_selection import cross_val_score
def score_dataset(X, y, model=XGBRegressor()):
    for colname in X.select_dtypes(["category"]):
        X[colname] = X[colname].cat.codes

    score_r2 = cross_val_score(model, X, y, cv=5, scoring="r2")
    score_r2 = score_r2.mean()

    return score_r2
```

The get_scores() function is defined to iterate over the models in the models dictionary, fit each model on the training data, and evaluate its performance using the score_dataset() function.

```python
 def get_scores(models,xtrain,ytrain):
    for name,model in models.items():
        model["model"].fit(xtrain,ytrain)

        score_r2 = score_dataset(xtrain, ytrain, model=model["model"])
        score = {'model':"Linear regression", 'score_r2':score_r2}
        print("--- "+name+" ---")
        print("Score r2: {}".format(score_r2))
        print("\n")
```
