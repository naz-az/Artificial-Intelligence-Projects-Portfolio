The next code block defines a function called split_data that splits the images from a source directory into training and validation sets. The function takes the source directory path, training directory path, validation directory path, and split size as inputs. It randomly shuffles the image files, splits them based on the split size, and copies the files to the appropriate directories.

```python
def split_data(SOURCE_DIR, TRAINING_DIR, VALIDATION_DIR, SPLIT_SIZE):
  """
  Splits the data into train and test sets

  Args:
    SOURCE_DIR (string): directory path containing the images
    TRAINING_DIR (string): directory path to be used for training
    VALIDATION_DIR (string): directory path to be used for validation
    SPLIT_SIZE (float): proportion of the dataset to be used for training

  Returns:
    None
  """
  all_files = os.listdir(SOURCE_DIR)
  valid_files = [f for f in all_files if os.path.getsize(os.path.join(SOURCE_DIR, f)) > 0]
  random.shuffle(valid_files)

  split_index = int(len(valid_files) * SPLIT_SIZE)
  train_files = valid_files[:split_index]
  val_files = valid_files[split_index:]

  for file in train_files:
      copyfile(os.path.join(SOURCE_DIR, file), os.path.join(TRAINING_DIR, file))

  for file in val_files:
      copyfile(os.path.join(SOURCE_DIR, file), os.path.join(VALIDATION_DIR, file))

```

The final code block defines a function called train_val_generators that creates the training and validation data generators using the ImageDataGenerator class from Keras. The function takes the training directory and validation directory as inputs and returns the train and validation generators. The generators apply data augmentation techniques such as rescaling, rotation, shifting, shearing, zooming, and flipping to the images.

```python
def train_val_generators(TRAINING_DIR, VALIDATION_DIR):
  """
  Creates the training and validation data generators

  Args:

    TRAINING_DIR (string): directory path containing the training images
    VALIDATION_DIR (string): directory path containing the testing/validation images

  Returns:
    train_generator, validation_generator - tuple containing the generators
  """

  # Instantiate the ImageDataGenerator class (don't forget to set the arguments to augment the images)
  train_datagen = ImageDataGenerator(rescale=1./255,
                                      rotation_range=40,
                                      width_shift_range=0.2,
                                      height_shift_range=0.2,
                                      shear_range=0.2,
                                      zoom_range=0.2,
                                      horizontal_flip=True,
                                      fill_mode='nearest')

  # Pass in the appropriate arguments to the flow_from_directory method
  train_generator = train_datagen.flow_from_directory(directory=TRAINING_DIR,
                                                      batch_size=100,
                                                      class_mode='binary',
                                                      target_size=(150, 150))

  # Instantiate the ImageDataGenerator class (don't forget to set the rescale argument)
  validation_datagen = ImageDataGenerator(rescale=1./255)

  # Pass in the appropriate arguments to the flow_from_directory method
  validation_generator = validation_datagen.flow_from_directory(directory=VALIDATION_DIR,
                                                                batch_size=100,
                                                                class_mode='binary',
                                                                target_size=(150, 150))
  return train_generator, validation_generator
```

```python
train_generator, validation_generator = train_val_generators(TRAINING_DIR, VALIDATION_DIR)
```

**Model: Convolutional**

The following code defines a function called create_model that builds a convolutional neural network (CNN) model for image classification. The model consists of several layers, including convolutional layers, max pooling layers, a flatten layer, and dense layers. The model is compiled with an optimizer, loss function, and evaluation metric.

```python
def create_model():

  tf.random.set_seed(42)

  model = tf.keras.models.Sequential([
      tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(512, activation='relu'),
      tf.keras.layers.Dense(1, activation='sigmoid')
  ])


  model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
                loss='binary_crossentropy',
                metrics=['accuracy'])


  return model
```

Next, the code block creates an instance of the model by calling the create_model function. This untrained model will be used for training in the subsequent steps. It then trains the model using the fit function, providing the training generator, number of epochs, verbosity, and validation data.

```python
# Get the untrained model
model = create_model()

# Train the model
history = model.fit(train_generator,
                    epochs=15,
                    verbose=1,
                    validation_data=validation_generator)
```

The next code block retrieves the accuracy, validation accuracy, loss, and validation loss values for each training epoch from the history object. These values will be used for plotting the training and validation curves.

```python
#-----------------------------------------------------------
# Retrieve a list of list results on training and test data
# sets for each training epoch
#-----------------------------------------------------------
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(len(acc)) # Get number of epochs

#------------------------------------------------
# Plot training and validation accuracy per epoch
#------------------------------------------------
plt.plot(epochs, acc, 'r', "Training Accuracy")
plt.plot(epochs, val_acc, 'b', "Validation Accuracy")
plt.title('Training and validation accuracy')
plt.show()
print("")

#------------------------------------------------
# Plot training and validation loss per epoch
#------------------------------------------------
plt.plot(epochs, loss, 'r', "Training Loss")
plt.plot(epochs, val_loss, 'b', "Validation Loss")
plt.show()
```

The function create_pre_trained_model initializes an InceptionV3 model using the specified local weights file. It loads the weights, makes all layers non-trainable, and returns the model.

```python
def create_pre_trained_model(local_weights_file):
  """
  Initializes an InceptionV3 model.

  Args:
    local_weights_file (string): path pointing to a pretrained weights H5 file

  Returns:
    pre_trained_model: the initialized InceptionV3 model
  """
  pre_trained_model = InceptionV3(input_shape = (150, 150, 3),
                                  include_top = False,
                                  weights = None)

  pre_trained_model.load_weights(local_weights_file)

  # Make all the layers in the pre-trained model non-trainable
  for layer in pre_trained_model.layers:
      layer.trainable = False

  return pre_trained_model
```

The custom model includes a flatten layer, a fully connected layer with 1024 hidden units and ReLU activation, a dropout layer, and a final sigmoid layer for classification. The complete model is compiled with the RMSprop optimizer, binary cross-entropy loss, and accuracy metric.

```python
def create_final_model(pre_trained_model, last_output):
  """
  Appends a custom model to a pre-trained model

  Args:
    pre_trained_model (tf.keras Model): model that will accept the train/test inputs
    last_output (tensor): last layer output of the pre-trained model

  Returns:
    model: the combined model
  """
  tf.random.set_seed(42)

  # Flatten the output layer to 1 dimension
  x = layers.Flatten()(last_output)

  # Add a fully connected layer with 1024 hidden units and ReLU activation
  x = layers.Dense(1024, activation='relu')(x)
  # Add a dropout rate of 0.2
  x = layers.Dropout(0.2)(x)
  # Add a final sigmoid layer for classification
  x = layers.Dense(1, activation='sigmoid')(x)

  # Create the complete model by using the Model class
  model = Model(inputs=pre_trained_model.input, outputs=x)

  # Compile the model
  model.compile(optimizer = RMSprop(learning_rate=0.0001),
                loss = 'binary_crossentropy',
                metrics = ['accuracy'])

  return model
```

The model is created by calling the create_final_model function with the pre_trained_model and last_output as inputs.

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import RMSprop

# Save your model in a variable
model = create_final_model(pre_trained_model, last_output)

# Inspect parameters
total_params = model.count_params()
num_trainable_params = sum([w.shape.num_elements() for w in model.trainable_weights])

print(f"There are {total_params:,} total parameters in this model.")
print(f"There are {num_trainable_params:,} trainable parameters in this model.")
```

```python
from tensorflow.keras.preprocessing import image

def display_random_predictions(model, class_names, n_images=15, ncols=3):
    plt.figure(figsize=(10, 10))

    validation_dirs = {
        "Cat": "/content/cats-v-dogs/validation/cats",
        "Dog": "/content/cats-v-dogs/validation/dogs"
    }

    for i in range(n_images):
        plt.subplot(5, ncols, i + 1)

        actual_label = random.choice(class_names)
        img_dir = validation_dirs[actual_label]

        random_img_filename = random.choice(os.listdir(img_dir))
        random_img_path = os.path.join(img_dir, random_img_filename)

        img = image.load_img(random_img_path, target_size=(150, 150))
        img_array = image.img_to_array(img) / 255.0
        plt.imshow(img_array)

        pred = model.predict(np.expand_dims(img_array, axis=0))[0][0]
        predicted_label = class_names[int(round(pred))]

        if actual_label == predicted_label:
            title_color = 'green'
        else:
            title_color = 'red'

        plt.title(f'Truth: {actual_label}\nPrediction: {predicted_label}', color=title_color)
        plt.axis(False)

    plt.show()

class_names = ['Cat', 'Dog']
display_random_predictions(model, class_names, n_images=15, ncols=3)
```
